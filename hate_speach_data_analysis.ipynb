```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_json("./data/found_tweets.txt", lines=True)
```

```python
df.groupby(df["user"].apply(lambda x: x["screen_name"]))["created_at"].count().sort_values(ascending=False).value_counts()
```

```python
(df.groupby(df["user"].apply(lambda x: x["screen_name"]))["created_at"].count() > 3).sum()
```

```python
df["created_at_hour"] = df["created_at"].dt.round("h")
df["screen_name"] = df["user"].apply(lambda x: x["screen_name"])
```

```python
df_by_time = df.groupby(["screen_name", "created_at_hour"])["created_at"].count().unstack().fillna(0)
df_by_time.head()
```
